```python
#Import numpy
import numpy as np

#Initialize variables
x = np.random.uniform(-22, 22)
print('Initial value of x :', x)
eta = 0.001
iterations = 1000

#Iteration
for _ in range(iterations):
    slope = 4 * x**3 + 6*x
    steplength = -eta * slope
    x += steplength
    
#Round x value and calculate objective function
x = round(x)
obj = x**4 + 3 * x**2 + 10

#Display values
print('Final value of x =', x)
print('Objective Function value =', obj)
```
